```python
''' to convert a normal number into Float and converting the numbers
 which contain "-" to their average '''

def convert(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
import numpy as np
def remove_psf_outliers(df):
    # dataframe as an input
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)# mean
        st = np.std(subdf.price_per_sqft)# standard deviation
        
        # keeping anything below mean - standard deviation and above mean + standard deviation in reduced data
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        #appending the dataframes per locations
        df_out = pd.concat([df_out, reduced_df],ignore_index=True)
    return df_out

df7 = remove_psf_outliers(df6)
df7.shape
```

```python
# creating a scatter plot for 2 and 3 bedrooms
import matplotlib
def plot_scatter(df,location):
    bhk2 = df[(df.location == location) & (df.bhk==2)]
    bhk3 = df[(df.location == location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', label='2 BHK',s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, color='red', label='3 BHK', marker='*',s=50)
    plt.xlabel('Total Square Feet Area')
    plt.ylabel('Price')
    plt.title(location)
    plt.legend()
    
plot_scatter(df7,"Hebbal")
# with this we get to know that there are 2 bedrooms with higher price then the 3 bedrooms
```

```python
# removing properties for same location, the price of 3 bedrooms apartment is less than that of 2 bedrooms with same sqft.
# removing those 2bhk apartment whose price_per_sqft is less than the mean price_per_sqft of 1bhk aparment 
# removing the outliers in bhk column
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
#we can't apply machine learning algo on charater columns to converting them to numerical using one hot encoder
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
# now to concat the df10 and dummies into a new dataframe
df11 = pd.concat([df10,dummies.drop('other', axis = 'columns')], axis='columns')
df11.head()
```

```python
df12 = df11.drop('location', axis='columns')
df12.head()
```

```python
# now since we are making price prediction model, so the dependent variable will be price 
x = df12.drop('price', axis='columns')
y = df12.price
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=10)
# random_state=10 is used to get the same training and test data each time when we run it
# test_size is used to specify proportion on data to be used in testing eg. 0.2 = 20%
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain, ytrain)
lr.score(xtest,ytest)
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit
# it will randomize sample so that each of the fold has equal distribution and not targeted into one area 

from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear_regression': {
            'model': Pipeline([
                ('scaler', StandardScaler()),  # Add scaling to the pipeline
                ('linear_regression', LinearRegression())
            ]),
            'params': {
                'linear_regression__fit_intercept': [True, False]  # Example parameter
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

```python
def predict_price(location,sqft,bath,bhk):
    global a
    location_index = np.where(x.columns==location)[0][0]
    
    a = np.zeros(len(x.columns))
    a[0] = sqft
    a[1] = bath
    a[2] = bhk
    if location_index >= 0:
        a[location_index] = 1
        
    return lr.predict([a])[0]
```

```python
predict_price('1st Phase JP Nagar', 1000, 2, 2)
```

```python
predict_price('1st Phase JP Nagar', 1000, 2, 3)
# here we can see that there are houses with 3 bedrooms whose prices are less than that of 2 bedrooms
# this is because that how our dataset contains data
```
